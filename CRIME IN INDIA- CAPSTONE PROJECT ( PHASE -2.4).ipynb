```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings # Filtering warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_excel("Phase 2 Crime data.xlsx")
#data taken from https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_crime_rate and census 2011
```

```python
data.drop(['Crime Rate 2018', 'Crime Rate 2019', 'Crime Rate 2020', 'Crime Rate 2021', 'Population', 'Literacy'], axis =1, inplace =True)
```

```python
data.head()
```

```python
columns = data.columns[1:]
columns
```

```python
for i in columns:
    plt.figure(figsize=(10, 5))
    sns.barplot(x='States',y=i, data=data)
    plt.xticks(rotation=90)
    plt.show()
```

```python
sns.barplot('States', 'Total Crimes 2018', color ='green', data=data)
plt.xticks(rotation =90)
ax2 = plt.twinx()
plt.plot('States', 'Area(Km2)', data=data, marker='.', color='red', label='Area')
plt.xlabel('States', fontsize=15)
plt.ylabel('Area(Km2)', fontsize=15)
plt.title('State Area vs Total Crime 2018')
plt.legend()
plt.show()
```

```python
sns.barplot('States', 'Total Crimes 2019', color ='red', data=data)
plt.xticks(rotation =90)
ax2 = plt.twinx()
plt.plot('States', 'Area(Km2)', data=data, marker='.', color='green', label='Area')
plt.xlabel('States', fontsize=15)
plt.ylabel('Area(Km2)', fontsize=15)
plt.title('State Area vs Total Crime 2019')
plt.legend()
plt.show()
```

```python
sns.barplot('States', 'Total Crimes 2020', color ='blue', data=data)
plt.xticks(rotation =90)
ax2 = plt.twinx()
plt.plot('States', 'Area(Km2)', data=data, marker='.', color='magenta', label='Area')
plt.xlabel('States', fontsize=15)
plt.ylabel('Area(Km2)', fontsize=15)
plt.title('State Area vs Total Crime 2020')
plt.legend()
plt.show()
```

```python
sns.barplot('States', 'Total Crimes 2021', color ='magenta', data=data)
plt.xticks(rotation =90)
ax2 = plt.twinx()
plt.plot('States', 'Area(Km2)', data=data, marker='.', color='blue', label='Area')
plt.xlabel('States', fontsize=15)
plt.ylabel('Area(Km2)', fontsize=15)
plt.title('State Area vs Total Crime 2021')
plt.legend()
plt.show()
```

```python
sns.barplot('States', 'Area(Km2)', color='yellow', data=data)
plt.xlabel('States', fontsize=15)
plt.ylabel('Area(Km2)', fontsize=15)
plt.xticks(rotation =90)
ax2 = plt.twinx()
plt.plot('States', 'Total Crimes 2018', color ='green', data=data, marker='^', label='2018')
plt.plot('States', 'Total Crimes 2019', color ='black', data=data, marker='.', label='2019')
plt.plot('States', 'Total Crimes 2020', color ='blue', data=data, marker='o', label='2020')
plt.plot('States', 'Total Crimes 2021', color ='red', data=data, marker='.', label='2021')
plt.title('State Area vs Total Crime 2018-2021')
plt.legend()
plt.show()
```
